**READING THE FILES AND VERIFYING IF EVERYTHING WORKS**

```python
from fitparse.base import FitFile
import sys
import fitparse as fit
from utils import project_root
import plotly.graph_objects as go

try:
    fitfile: FitFile = fit.FitFile(f"{project_root}/training_data/user1/25.06.20-fit.fit")
    fitfile.parse()
except FitParseError as e:
    print (f"Error while parsing .FIT file: {e}")
    sys.exit(1)

fields = set()
for record in fitfile.get_messages('record'):
    for data in record: 
        fields.add(data.name)
print(fields)

fields_to_select = ["heart_rate", "power", "cadence", "speed", "altitude"]
series = {f: {"t": [], "v": []} for f in fields}
for record in fitfile.get_messages("record"):
    values = record.get_values()
    ts = values.get("timestamp")
    for f in fields_to_select:
        if values.get(f) is not None:
            series[f]["t"].append(ts)
            series[f]["v"].append(values[f])

fig = go.Figure()
for f in fields:
    fig.add_scatter(x=series[f]["t"], y=series[f]["v"], name=f, mode="lines")

fig.update_layout(hovermode="x unified", title="FIT record data")
fig.show()
```
